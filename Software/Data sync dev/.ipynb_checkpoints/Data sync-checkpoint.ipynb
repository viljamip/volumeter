```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from math import ceil
%matplotlib inline
```

```python
def smooth(df,window_len=11,window='hanning'):
    x = df.iloc[:,0]
    if x.ndim != 1:
        raise ValueError("smooth only accepts 1 dimension arrays.")
        
    if window_len % 2 != 1:
        raise ValueError("use an odd integer for the window_len.")
        
    if x.size < window_len:
        raise ValueError("Input vector needs to be bigger than window size.")

    if window_len<3:
        return x

    if not window in ['flat', 'hanning', 'hamming', 'bartlett', 'blackman']:
        raise ValueError("Window is on of 'flat', 'hanning', 'hamming', 'bartlett', 'blackman'")

    
    s=np.r_[x[window_len-1:0:-1],x,x[-2:-window_len-1:-1]]
    #print(len(s))
    if window == 'flat': #moving average
        w=np.ones(window_len,'d')
    else:
        w=eval('np.'+window+'(window_len)')

    y=np.convolve(w/w.sum(),s,mode='valid')
    #return y
    #return y[(ceil(window_len/2)-1):-(ceil(window_len/2))]
    dfSmooth = pd.DataFrame()
    dfSmooth['smooth'] = y[(ceil(window_len/2)-1):-(ceil(window_len/2))]
    dfSmooth.index = df.index
    return dfSmooth
```

```python
def rms(x):
    return np.sqrt(x.dot(x)/x.size)
```

```python
initialWeight = 8625.323863867425
holderData = pd.read_pickle('specimenHolder')
measurementData = pd.read_pickle('measurement')
measurementData -= initialWeight

measurementData = smooth(measurementData, 77)
holderData = smooth(holderData, 77)
diff = measurementData.iloc[:,0]-holderData.iloc[:,0]

# DEBUG
measurementData = measurementData.head(1000)
holderData = holderData.head(1000)
diff = diff.head(1000)
```

Eli etsitään ensimmäinen arvo, joka on alle thresholdin, sitten etsitään koska arvo alkaa taas kasvaa. Minimikohta on tätä edeltävä arvo
   

```python
def findSyncPoint(data):
    threshold = -10 * rms(data.iloc[:,0].head(200))
    
    index = 0
    
    while(index < len(data.iloc[:,0])):
        val = data.iloc[:,0][index]
        if(val <= threshold):
            prevVal = val
            index += 1
            val = data.iloc[:,0][index]
            while(val < prevVal):
                index += 1
                prevVal = val
                val = data.iloc[:,0][index]
                
            return index - 1
        
        index += 1
    return None
```
